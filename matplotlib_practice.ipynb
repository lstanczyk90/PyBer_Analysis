```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.

plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt().

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="green", linewidth=2, marker="D", label="Boston")
ax.set_title("PyBer Far By Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0, 45)
ax.grid()
ax.legend()
```

```python
# Create the plot and add a label for the legend.

plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

# Create labels for the x and y axes

plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45

plt.ylim(0, 45)

# Create a Title.

plt.title("PyBer Fare By Month")

# Add a grid

plt.grid()

# Add the legend.

plt.legend()
```

```python
# Create the Bar Chart

plt.bar(x_axis, y_axis, color="green", label="Boston")

# Create labels for the x and y axis.

plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a Title.

plt.title("PyBer Fare By Month")

# Add the legend.

plt.legend()
```

```python
# Create the plot.
# NOTE: GCA means get current axis; we are chaining with invert method.

plt.barh(x_axis, y_axis, color="purple", label="Boston")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_title("PyBer Far By Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.legend()
plt.gca().invert_yaxis()
```

```python
# Set the x-axis to a list of strings for each month.
x_values = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_values = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

fig, ax = plt.subplots()
ax.scatter(y_values, x_values, label = "Boston", facecolor = "skyblue", edgecolor = "black", s = [(data*5) for data in y_values], alpha = 0.2, linewidth = 2)

ax.set_title("PyBer Fare By Month")
ax.set_xlim(0,50,10)
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.legend()
plt.gca().invert_yaxis()
```
